```python
import pandas as pd
```

```python
df=pd.read_csv("Airbnb_Open_Data.csv",low_memory=False)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.drop(columns=['license'], inplace=True)
```

```python
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(' ','_')
)
```

```python
df['price'].head()
```

```python
df['price'] = (
    df['price']
      .str.replace('$','',regex=False)
      .str.replace(',','',regex=False)
      .astype(float)
)

df['service_fee'] = (
    df['service_fee']
      .str.replace('$','',regex=False)
      .str.replace(',','',regex=False)
      .astype(float)
)
```

```python
df['last_review'] = pd.to_datetime(
    df['last_review'],
    errors='coerce'
)
```

```python
nulls = (
    df.isnull().sum()
      .sort_values(ascending=False)
)
```

```python
df['house_rules'] = df['house_rules'].fillna('Not Provided')
```

```python
df['host_identity_verified'] = (
    df['host_identity_verified']
      .fillna('unconfirmed')
)
```

```python
df['country'].value_counts()
```

```python
df['country'] = (
    df['country']
      .fillna('United States')
)
```

```python
# df[df['minimum_nights'] < 0]
```

```python
df = df[df['minimum_nights'] >= 0]
```

```python
df = df[
    (df['availability_365'] >= 0) &
    (df['availability_365'] <= 365)
]
```

```python
df['neighbourhood_group'] = (
    df['neighbourhood_group']
      .str.strip()
      .str.title()
)
```

```python
df['reviews_per_month'] = (
    df['reviews_per_month']
      .fillna(df['reviews_per_month'].median())
)
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.to_csv(
    'Airbnb_Cleaned.csv',
    index=False
)
```
